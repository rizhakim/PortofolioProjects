```python
import pandas as pd
import numpy as np
import pathlib
import os
```

```python
#Decide folder path
folder_path = "E:/Temporary/NCR/"

#Make list of file
files = os.listdir(folder_path)

#Make list of file with folder path
excel_list = []
for file in files:
    excel_list.append(folder_path+file)
    
#Initialize an empty DataFrame for Combining Excel
df_combine = pd.DataFrame()

#Using for loop to cleaning each excel file then combining
for file in excel_list:
    
    #import excel file, set header as number
    df = pd.read_excel(file, header= None)
    
    #delete null (NaN) rows based on column no. 7 (column no.7 must be filled in this case)
    df.drop(df[df[7].isnull()].index, inplace = True)
    
    #reset index number
    df = df.reset_index(drop=True)

    #Set column name to the first row
    df.columns = df.iloc[0,:]

    #delete row not header, first row still present
    df.drop(df[df['NO']=='NO'].index, inplace = True)

    #selecting important columns 
    df = df[['NO','DATE','CLASS','SIZE','PRODUK', 'MC', 'PROBLEM', 'REASON']]

    #Concat or adding each result of dataframe to combined dataframe
    df_combine = df_combine.append(df)
    
df_combine
```

```python
#exporting to excel file
df_combine.to_excel(r"E:\Temporary\NCR\Combined NCR Data.xlsx")
```
